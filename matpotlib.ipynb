```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Create figure and axis
fig, ax = plt.subplots()
x_data, y_data = [], []
line, = ax.plot([], [], "r-o")  # Red line with circle markers

# Set limits
ax.set_xlim(0, 10)
ax.set_ylim(0, 100)

# Update function for animation
def update(frame):
    x_data.append(frame)
    y_data.append(np.random.randint(0, 100))  # Random Y values
    line.set_data(x_data, y_data)
    return line,

# Create animation
ani = animation.FuncAnimation(fig, update, frames=np.arange(0, 11, 1), interval=500)

# Show animation
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Create figure and axis
fig, ax = plt.subplots()
x_labels = ["A", "B", "C", "D", "E"]
y_values = np.random.randint(10, 100, size=5)
bars = ax.bar(x_labels, y_values, color="skyblue")

# Update function
def update(frame):
    new_values = np.random.randint(10, 100, size=5)  # Generate new data
    for bar, new_value in zip(bars, new_values):
        bar.set_height(new_value)  # Update bar heights

# Create animation
ani = animation.FuncAnimation(fig, update, frames=20, interval=500)

# Show animation
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# Create figure and axis
fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.25)  # Space for slider

# Generate initial sine wave
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
line, = ax.plot(x, y, 'b-')

# Add slider axis
ax_slider = plt.axes([0.2, 0.1, 0.65, 0.03])  # [left, bottom, width, height]
slider = Slider(ax_slider, "Frequency", 0.1, 5.0, valinit=1.0)

# Update function
def update(val):
    freq = slider.val
    line.set_ydata(np.sin(freq * x))  # Update sine wave
    fig.canvas.draw_idle()  # Refresh plot

# Connect slider to function
slider.on_changed(update)

# Show interactive plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import RadioButtons

# Create figure and axis
fig, ax = plt.subplots()
plt.subplots_adjust(left=0.3)

# Generate initial data
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
line, = ax.plot(x, y, 'b-')

# Dropdown (Radio Buttons)
ax_radio = plt.axes([0.05, 0.4, 0.2, 0.2])  # Position for radio buttons
radio = RadioButtons(ax_radio, ('Sine', 'Cosine', 'Tangent'))

# Update function
def update(label):
    if label == "Sine":
        line.set_ydata(np.sin(x))
    elif label == "Cosine":
        line.set_ydata(np.cos(x))
    elif label == "Tangent":
        line.set_ydata(np.tan(x))
    fig.canvas.draw_idle()

# Connect radio buttons to function
radio.on_clicked(update)

# Show interactive plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider, Button, RadioButtons

# Create figure and axis
fig, ax = plt.subplots()
plt.subplots_adjust(left=0.3, bottom=0.3)

# Generate initial sine wave
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
line, = ax.plot(x, y, 'b-')

# Sliders for amplitude and frequency
ax_amp = plt.axes([0.2, 0.15, 0.65, 0.03])
slider_amp = Slider(ax_amp, "Amplitude", 0.1, 2.0, valinit=1.0)

ax_freq = plt.axes([0.2, 0.1, 0.65, 0.03])
slider_freq = Slider(ax_freq, "Frequency", 0.1, 5.0, valinit=1.0)

# Button to reset sliders
ax_button = plt.axes([0.8, 0.05, 0.1, 0.075])
button = Button(ax_button, "Reset")

# Radio buttons to change wave type
ax_radio = plt.axes([0.05, 0.4, 0.2, 0.2])
radio = RadioButtons(ax_radio, ('Sine', 'Cosine', 'Tangent'))

# Update function
def update(val):
    amp = slider_amp.val
    freq = slider_freq.val
    wave_type = radio.value_selected  # Get selected wave
    if wave_type == "Sine":
        line.set_ydata(amp * np.sin(freq * x))
    elif wave_type == "Cosine":
        line.set_ydata(amp * np.cos(freq * x))
    elif wave_type == "Tangent":
        line.set_ydata(amp * np.tan(freq * x))
    fig.canvas.draw_idle()

# Reset function
def reset(event):
    slider_amp.reset()
    slider_freq.reset()

# Connect widgets
slider_amp.on_changed(update)
slider_freq.on_changed(update)
radio.on_clicked(update)
button.on_clicked(reset)

# Show interactive plot
plt.show()
```

```python
import tkinter as tk
from tkinter import ttk
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

# Create main window
root = tk.Tk()
root.title("Matplotlib in Tkinter")
root.geometry("600x500")

# Create a Matplotlib figure
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4, 5], [10, 20, 25, 30, 40], marker='o', linestyle='-')
ax.set_title("Simple Line Plot")

# Embed Matplotlib figure into Tkinter
canvas = FigureCanvasTkAgg(fig, root)
canvas.get_tk_widget().pack(expand=True, fill=tk.BOTH)
canvas.draw()

# Run Tkinter event loop
root.mainloop()
```
