```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./dataset/Unicorn_Companies.csv")
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.sample()
```

```python
df["Founded Year"].fillna(df["Founded Year"].mode()[0], inplace=True)
df["Total Raised"].fillna(df["Total Raised"].median(), inplace=True)
df["Investors Count"].fillna(df["Investors Count"].median(), inplace=True)
df["Deal Terms"].fillna(df["Deal Terms"].median(), inplace=True)
df["Portfolio Exits"].fillna(df["Portfolio Exits"].median(), inplace=True)
df["Financial Stage"].fillna(df["Financial Stage"].mode()[0], inplace=True)
df["Select Inverstors"].fillna("Unknown", inplace=True)
```

```python
df.sample()
```

```python
df.info()
```

```python
top_countries=df['Country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette="viridis")
plt.xlabel("Number of Unicorn Companies")
plt.ylabel("Country")
plt.title("Top 10 Countries with the Most Unicorn Companies")
plt.show()
```

```python
founded_counts = df["Founded Year"].value_counts().sort_index()
sns.lineplot(x=founded_counts.index, y=founded_counts.values, marker="o", color="red")
plt.xlabel("Founded Year")
plt.ylabel("Number of Unicorns")
plt.title("Number of Unicorns Founded Each Year")
plt.show()
```

```python
industry_counts = df["Industry"].value_counts().head(5)
plt.pie(industry_counts, labels=industry_counts.index, autopct="%1.1f%%", colors=sns.color_palette("pastel"))
plt.title("Top 5 Industries of Unicorn Companies")
plt.show()
```

```python
top_cities = df["City"].value_counts().head(10)
sns.barplot(x=top_cities.values, y=top_cities.index, palette="magma")
plt.xlabel("Number of Unicorn Companies")
plt.ylabel("City")
plt.title("Top 10 Cities with the Most Unicorn Companies")
plt.show()
```

```python
investor_list = df["Select Inverstors"].str.split(", ").explode()  # Split multiple investors per row
top_investors = investor_list.value_counts().head(10)
sns.barplot(x=top_investors.values, y=top_investors.index, palette="crest")
plt.xlabel("Number of Unicorn Companies Invested In")
plt.ylabel("Investor")
plt.title("Top 10 Investors by Number of Unicorns Invested In")
plt.show()
```
