```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\borud\\OneDrive\\Desktop\\matpotlit\\dataset\\university_admission.csv")
df
```

```python
# Donut Chart - Research Distribution
research_counts = df["Research"].value_counts()
labels = ["No Research", "Has Research"]
colors = ["#ff9999", "#66b3ff"]

plt.figure(figsize=(8, 8))
plt.pie(research_counts, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors, wedgeprops={'edgecolor': 'black'})

# Add a white circle to create the donut effect
center_circle = plt.Circle((0, 0), 0.6, color='white')
plt.gca().add_artist(center_circle)
plt.title("Distribution of Candidates Based on Research")
plt.show()
```

```python
x = df.index 
gre_scores = df["GRE Score"]
toefl_scores = df["TOEFL Score"] - df["GRE Score"] 
cgpa_scores = df["CGPA"] - df["TOEFL Score"]
plt.figure(figsize=(12, 6))
plt.stackplot(x, gre_scores, toefl_scores, cgpa_scores, labels=["GRE Score", "TOEFL Score", "CGPA"], 
              colors=["#ff9999", "#66b3ff", "#99cc99"], alpha=0.7)
plt.xlabel("Candidate Index")
plt.ylabel("Scores")
plt.title("Stacked Area Chart of GRE, TOEFL, and CGPA Trends")
plt.legend(loc="upper left")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
