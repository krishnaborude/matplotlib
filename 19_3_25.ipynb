```python
cancel_counts = df["is_canceled"].value_counts()

plt.bar(["Not Canceled", "Canceled"], cancel_counts, color=["blue", "red"])

plt.xlabel("Booking Status")
plt.ylabel("Count")
plt.title("Hotel Booking Cancellations")

plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(cancel_counts,labels=["Not Canceled", "Canceled"],colors=["b","r"],autopct="%1.2f%%",startangle=90)
plt.title("Hotel Booking Cancellations")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["lead_time"], df["adr"], color="blue", alpha=0.5)
plt.xlabel("Lead Time (Days)")
plt.ylabel("Average Daily Rate (ADR)")
plt.title("Scatter Plot: Lead Time vs. ADR")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
yearly_bookings = df["arrival_date_year"].value_counts().sort_index()
plt.bar(yearly_bookings.index, yearly_bookings.values, color="blue")
plt.xlabel("Year")
plt.ylabel("Number of Bookings")
plt.title("Number of Bookings per Year")
plt.xticks(yearly_bookings.index) 
plt.show()
```

```python
plt.figure(figsize=(8, 5))
monthly_bookings = df["arrival_date_month"].value_counts().sort_index()
plt.plot(monthly_bookings.index, monthly_bookings.values, marker="o", linestyle="-", color="blue")
plt.xlabel("Month")
plt.ylabel("Number of Bookings")
plt.title("Number of Bookings per Month")
plt.xticks(rotation=45)  
plt.show()
```

```python
plt.figure(figsize=(12, 8))
day_bookings = df["arrival_date_day_of_month"].value_counts().sort_index()
plt.bar(day_bookings.index, day_bookings.values, color="blue")
plt.xlabel("Day of the Month")
plt.ylabel("Number of Bookings")
plt.title("Number of Bookings per Day of the Month")
plt.xticks(day_bookings.index) 
plt.show()
```
