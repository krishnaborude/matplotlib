```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(2000, 2025)  
category1 = np.random.randint(50, 100, len(x))
category2 = np.random.randint(30, 80, len(x))  
category3 = np.random.randint(20, 60, len(x))

plt.stackplot(x, category1, category2, category3, labels=['A', 'B', 'C'],alpha=0.5,)

plt.xlabel("Year")
plt.ylabel("Value")
plt.title("Stacked Area Chart Example")
plt.legend(loc='upper left', fontsize=10, shadow=True, facecolor='white', edgecolor='black')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
labels = ['Apple', 'Banana', 'Cherry', 'Mango']
sizes = [25, 35, 30, 10]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
explode = (0, 0.1, 0, 0)
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.0f%%', startangle=140,explode=explode,shadow=True)
center_circle = plt.Circle((0,0), 0.5, color='white')
plt.text(0, 0, 'Fruits', fontsize=14, ha='center', va='center', fontweight='bold')
plt.gca().add_artist(center_circle)
plt.axis('equal')
plt.legend(loc="upper right", bbox_to_anchor=(1.2, 1))
plt.title("Donut Chart Example")
plt.show()
```
