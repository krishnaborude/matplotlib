```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(2000, 2025)  # Years
category1 = np.random.randint(50, 100, len(x))  # Category A
category2 = np.random.randint(30, 80, len(x))   # Category B
category3 = np.random.randint(20, 60, len(x))   # Category C

# Stack the data
plt.stackplot(x, category1, category2, category3, labels=['A', 'B', 'C'],alpha=0.5,)

# Labels and legend
plt.xlabel("Year")
plt.ylabel("Value")
plt.title("Stacked Area Chart Example")
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [25, 35, 30, 10]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

# Create a pie chart (donut chart)
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Add a white circle to create the donut effect
center_circle = plt.Circle((0,0), 0.5, color='white')
plt.gca().add_artist(center_circle)

# Set aspect ratio to be equal (circular)
plt.axis('equal')

# Add title
plt.title("Donut Chart Example")

# Show the chart
plt.show()
```
