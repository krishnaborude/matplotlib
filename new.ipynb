```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,5,43,3]
y=[10,20,30,40,50,35,35,23]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,c='g',marker='o',markersize=10, linestyle=':')
plt.ylabel('value')
plt.xlabel('number')
plt.title('line chart')
plt.show()
```

```python
plt.ylabel('value')
plt.xlabel('number')
plt.grid()
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
k=[1,3,1,4,2,5,8]
l=[3,2,5,6,6,44,3]
plt.scatter(k,l,c='r')

plt.scatter(x,y,c='g')
plt.show()
```

```python
color=["red","green","blue","yellow","pink","black","orange","purple"]
size=[10,20,30,40,50,60,70,80]
plt.scatter(x,y,c=color,s=size,alpha=0.5)
```

```python
Name=['krishna','ajinkya','prasad','shubham']
mark=[100,30,50,20]
plt.title('student mark bar chart',loc='center',size=20,color='hotpink')
plt.ylabel('mark of student',color='red')
plt.xlabel('name of student',color='red')
plt.bar(Name,mark,width=0.5,color='g')
plt.show()
```

```python
Name=['krishna','ajinkya','prasad','shubham']
mark=[100,30,50,20]
plt.title('student mark bar chart',loc='center',size=20,color='hotpink')
plt.ylabel('mark of student',color='red')
plt.xlabel('name of student',color='red')
plt.barh(Name,mark,color='pink',height=0.5)
plt.show()
```
